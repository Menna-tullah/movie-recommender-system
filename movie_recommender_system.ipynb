```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#get the data
column_names = ['user_id', 'item_id', 'rating', 'timestamp']
df = pd.read_csv('u.data', sep='\t', names=column_names)
```

```python
df.head()
```

```python
movie_titles = pd.read_csv('Movie_Id_Titles')
movie_titles.head()
```

```python
df = pd.merge(df, movie_titles, on='item_id')
df.head()
```

```python
sns.set_style('white')
```

```python
df.groupby('title')['rating'].mean().sort_values(ascending=False).head()
```

```python
df.groupby('title')['rating'].count().sort_values(ascending=False).head()
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings.head()
```

```python
#to set the column of number of ratings
ratings['number_of_ratings'] = pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head()
```

```python
plt.figure(figsize=(10,6))
ratings['number_of_ratings'].hist(bins=70)
```

```python
plt.figure(figsize=(10,6))
ratings['rating'].hist(bins=70)
```

```python
sns.jointplot(x="rating", y="number_of_ratings", data=ratings, alpha=0.5)
```

```python
#create matrix which include the column of the user_id and each user ratings for various movies
moviemat = df.pivot_table(index='user_id', columns='title', values='rating')
moviemat.head()
```

```python
ratings.sort_values('number_of_ratings', ascending=False).head(10)
```

```python
ratings.head()
```

```python
#picking two movies to use which are star wars, Liar Liar
starwars_user_ratings = moviemat['Star Wars (1977)']
liarliar_user_ratings = moviemat['Liar Liar (1997)']
starwars_user_ratings.head()
```

```python
similar_to_starwars = moviemat.corrwith(starwars_user_ratings)
similar_to_liarliar = moviemat.corrwith(liarliar_user_ratings)
```

```python
#removing NAN values and work with dataframe instead of series
corr_starwars = pd.DataFrame(similar_to_starwars, columns=['correlation'])
corr_starwars.dropna(inplace=True)
corr_starwars.head()
```

```python
#to get the most similar movies
corr_starwars.sort_values('correlation', ascending=False).head(10)
```

```python
#filtring the movies that has less than 100 number of ratings to enhance the results
corr_starwars = corr_starwars.join(ratings['number_of_ratings'])
corr_starwars.head()
```

```python
#drop the ratings of movies with less 100 number of ratings
corr_starwars[corr_starwars['number_of_ratings']>100].sort_values('correlation', ascending=False).head()
```

```python
#same for Liar Liar movie
corr_liarliar = pd.DataFrame(similar_to_liarliar, columns=['correlation'])
corr_liarliar.dropna(inplace=True)
corr_liarliar = corr_liarliar.join(ratings['number_of_ratings'])
corr_liarliar[corr_liarliar['number_of_ratings']>100].sort_values('correlation', ascending=False).head()
```
